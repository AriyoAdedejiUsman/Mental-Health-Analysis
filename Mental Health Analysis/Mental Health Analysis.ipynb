```python
#importing important libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Data preparation and Cleaning

```python
#loading data
data = pd.read_csv('Mental Health Dataset.csv')
```

```python
data.isnull().sum()
```

```python
# Impute missing values with mode
data['self_employed'].fillna(data['self_employed'].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

#  Data Visualisation

# Distribution of Gender

```python
ax = sns.countplot(x='Gender', data=data)
plt.title('Distribution of Gender')

for bars in ax.containers:
    ax.bar_label(bars)


plt.show()

```

# Count plot of Treatment by Gender

```python
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='treatment', hue='Gender', data=data)
plt.title('Distribution of Treatment by Gender')
plt.xlabel('Treatment Received')
plt.ylabel('Count')
plt.legend(title='Gender')


for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

The visualizations above illustrate a notable gender disparity in mental health issues, with a higher prevalence observed among males compared to females. Furthermore, a significant proportion across both genders do not receive treatment. Interestingly, despite the smaller population of females with mental health issues, a majority of them seek treatment, suggesting a proactive approach to mental health care among women

# Distribution of occupation

```python
ax = sns.countplot(x='Occupation', data=data)
plt.title('Distribution of Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Count plot of self_employed
plt.figure(figsize=(8, 6))
sns.countplot(x='self_employed', data=data)
plt.title('Distribution of Self-Employment Status')
plt.xlabel('Self-Employed')
plt.ylabel('Count')
plt.show()
```

# Distribution of Country

```python
plt.figure(figsize=(14, 6))
ax = sns.countplot(y='Country', data=data, order=data['Country'].value_counts().index)
plt.title('Distribution of Country')


for bars in ax.containers:
    ax.bar_label(bars)


plt.show()
```

The visualizations highlight that the United States, the United Kingdom, and Canada exhibit the highest number of reported mental health cases. Notably, individuals in various occupations, including housewives, students, and corporate workers, comprise a significant portion of these cases. Additionally, a substantial number of individuals who are not self-employed are affected by mental health issues, possibly indicating a correlation with work-related stress or pressure.

# Relationship between Mood_Swings and Treatment

```python
ax = sns.countplot(x='Mood_Swings', hue='treatment', data=data)
plt.title('Relationship between Mood Swings and Treatment')


for bars in ax.containers:
    ax.bar_label(bars)
    
plt.show()
```

# Visualization of Symptom Profiles

```python
symptoms = ['Mood_Swings', 'Coping_Struggles', 'Social_Weakness']
plt.figure(figsize=(10, 6))
for i, symptom in enumerate(symptoms, 1):
    plt.subplot(1, len(symptoms), i)
    ax=sns.countplot(x=symptom, data=data)
    plt.title(f'Distribution of {symptom}')
plt.tight_layout()

for bars in ax.containers:
    ax.bar_label(bars)


plt.show()
```

The visualizations above reveal that a majority of individuals exhibit moderate symptoms, report no coping struggles, and are uncertain about their social weaknesses.

# Access to Care Options count plot

```python
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='care_options', data=data)
plt.title('Access to Care Options')
plt.xlabel('Care Options')
plt.ylabel('Count')

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

# Count plot of Days Indoors

```python
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='Days_Indoors', data=data)
plt.title('Distribution of Days Indoors')
plt.xlabel('Days Indoors')
plt.ylabel('Count')
plt.xticks(rotation=45)

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

Based on the above visualizations, it's noteworthy that there is no substantial variance in the distribution of "Days_Indoor." The highest frequency is observed within the range of 1-14 days, followed by 31-60 days. Furthermore, the analysis of "care_options" indicates that the majority of individuals do not have any options in place.

# Count plot of Family History

```python
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='family_history', data=data)
plt.title('Distribution of Family History of Mental Health Issues')
plt.xlabel('Family History')
plt.ylabel('Count')

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

This visualization illustrates that the number of individuals with a family history of mental health issues is significantly lower compared to those without such a history within our dataset. 

# Conclusion

"After analyzing the various visualizations, several key insights emerge regarding mental health trends within our dataset. Firstly, there is a notable gender disparity, with males exhibiting a higher prevalence of mental health issues compared to females. Despite this, a considerable proportion of both genders do not receive treatment, highlighting potential gaps in mental health care access. Additionally, the United States, the United Kingdom, and Canada stand out as countries with the highest incidence of mental health cases, with housewives, students, and corporate workers being the most affected occupational groups.

Regarding symptoms, a majority of individuals exhibit moderate symptomatology and report no coping struggles. However, uncertainty regarding social weaknesses is prevalent among respondents. Interestingly, while the distribution of "Days_Indoor" does not exhibit significant variance, the majority of individuals spend between 1-14 days indoors, followed by 31-60 days.

In addition to the aforementioned observations, the visualization highlights a notable trend: the prevalence of individuals with a family history of mental health issues is significantly lower compared to those without such a history in our dataset. This suggests that environmental factors and other influences may play a substantial role in the manifestation of mental health issues.

Moreover, concerning care options, the majority of individuals do not have any options in place, indicating potential gaps in mental health care infrastructure. In summary, these observations underscore the need for targeted interventions to address disparities in treatment access, enhance mental health support infrastructure, and promote proactive approaches to mental well-being across different demographics."
